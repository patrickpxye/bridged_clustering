```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D, art3d
import matplotlib.patches as patches
import numpy as np

def draw_3d_pie(ax, center, radius, proportions, colors, labels=None):
    """
    Draws a pie chart on a 3D axis by creating 2D wedge patches and converting them.
    
    Parameters:
        ax: The 3D axes to draw on.
        center: Tuple (x, y, z) indicating the center position of the pie.
        radius: Radius of the pie.
        proportions: List of proportions (should sum to 1).
        colors: List of colors for each wedge.
        labels: Optional list of labels for each wedge.
    """
    start_angle = 0
    for i, prop in enumerate(proportions):
        end_angle = start_angle + prop * 360
        # Create a 2D wedge
        wedge = patches.Wedge(
            center=(center[0], center[1]), 
            r=radius,
            theta1=start_angle,
            theta2=end_angle,
            facecolor=colors[i],
            edgecolor='k',
            linewidth=0.1
        )
        # Convert to 3D
        ax.add_patch(wedge)
        art3d.pathpatch_2d_to_3d(wedge, z=center[2], zdir="z")
        
        # Place label text with a higher z-offset
        if labels:
            angle_mid = np.deg2rad((start_angle + end_angle) / 2)
        start_angle = end_angle

# Data
sample_sizes = [30, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130]
model1_probs = [0.5, 0.4, 0,0, 0.3, 0.2, 0.1, 0.2, 0.1, 0.1, 0.3]  # Example probabilities for Model 1
model2_probs = [1 - p for p in model1_probs]

# Colors and labels for wedges
colors = ['blue', 'orange']
pie_labels = ['Model 1', 'Model 2']

# Set up 3D figure
fig = plt.figure(figsize=(8, 14))
ax = fig.add_subplot(111, projection='3d')

radius = 3
spacing = 5

for i, size in enumerate(sample_sizes):
    center = (i * spacing, 0, 0)  # place pies along x-axis
    proportions = [model1_probs[i], model2_probs[i]]
    
    draw_3d_pie(ax, center, radius, proportions, colors, labels=pie_labels)
    
    # Label each pie with sample size, also with a slight z-offset
    ax.text(
        center[0] + 1.0 , center[1], 0.5, 
        f"n={size}",
        ha='center',
        va='center',
        fontsize=10,
        color='black',
        zorder=10
    )

# Adjust the viewing angle and axis limits
ax.view_init(elev=30, azim=-60)  # tilt downward, rotate for a better look
ax.set_xlim(-1, len(sample_sizes) * spacing + 1)
ax.set_ylim(-radius - 2, radius + 2)
ax.set_zlim(0, 2)

# Optionally hide axes for a cleaner look
ax.set_axis_off()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Given data
sample_sizes = [30, 40, 50, 60, 70, 80, 90, 100, 110, 120, 130]
model1_probs = [0.5, 0.4, 0.0, 0.0, 0.3, 0.2, 0.1, 0.2, 0.1, 0.1, 0.3]

# Convert probabilities to percentages
model1_percentages = [prob * 100 for prob in model1_probs]
complement_percentages = [100 - p for p in model1_percentages]

# Plotting the bar plot with complementary bars
plt.figure(figsize=(8, 5))

# Plot the original probabilities as filled bars
plt.bar(sample_sizes, model1_percentages, width=5, label='BKM success', color='blue')

# Plot the complementary probabilities as stacked bars
plt.bar(sample_sizes, complement_percentages, width=5, bottom=model1_percentages, label='AGDN success', color='orange')

plt.xlabel('Sample Sizes')
plt.ylabel('Percentage')
plt.legend()
plt.show()
```
